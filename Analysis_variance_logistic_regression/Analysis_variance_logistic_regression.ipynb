### Провести дисперсионный анализ для определения того, есть ли различия среднего роста среди взрослых футболистов, хоккеистов и штангистов. Даны значения роста в трех группах случайно выбранных спортсменов: Футболисты: 173, 175, 180, 178, 177, 185, 183, 182. Хоккеисты: 177, 179, 180, 188, 177, 172, 171, 184, 180. Штангисты: 172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

football_players = [173, 175, 180, 178, 177, 185, 183, 182]
f = [1 for i in range(len(football_players))]
football_players_mean = np.mean(football_players)

hockey_players = [177, 179, 180, 188, 177, 172, 171, 184, 180]
h = [2 for i in range(len(hockey_players))]
hockey_players_mean = np.mean(hockey_players)

weightlifters = [172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170]
w = [3 for i in range(len(weightlifters))]
weightlifters_mean = np.mean(weightlifters)


mean_all = sum([*football_players, *hockey_players, *weightlifters]) / (len(football_players) 
                                                                   + len(hockey_players) 
                                                                   + len(weightlifters))

plt.scatter(f, football_players, label = 'football_players')
plt.scatter(1.25, football_players_mean, label = 'football_players_mean',)

plt.scatter(h, hockey_players, label = 'hockey_players')
plt.scatter(2.25, hockey_players_mean, label = 'hockey_players_mean')


plt.scatter(w, weightlifters, label = 'weightlifters')
plt.scatter(3.25, weightlifters_mean, label = 'weightlifters_mean')

plt.scatter(3.5, mean_all, label = 'mean_all')

plt.legend(loc='upper right', fontsize = 8)
plt.show()
```

```python
mean_all = sum([*football_players, *hockey_players, *weightlifters]) / len([*football_players, *hockey_players, *weightlifters])
```

```python
sum_squared_deviations = sum((np.array(football_players) - mean_all) ** 2) + sum((np.array(hockey_players) - mean_all) ** 2) + sum((np.array(weightlifters) - mean_all) ** 2)
print(sum_squared_deviations)
```

```python
s_fact = ((football_players_mean - mean_all) ** 2) * len(football_players) + ((hockey_players_mean - mean_all) ** 2) * len(hockey_players) + ((weightlifters_mean - mean_all) ** 2) * len(weightlifters)
print(s_fact)

s_ost = sum((np.array(football_players) - football_players_mean) ** 2) + sum((np.array(hockey_players) - hockey_players_mean) ** 2) + sum((np.array(weightlifters) - weightlifters_mean) ** 2)
print(s_ost)
print(s_fact + s_ost)
```

```python
sigma_fact = s_fact / (3 -1)
print(sigma_fact)
```

```python
sigma_ost = s_ost / (len([*football_players, *hockey_players, *weightlifters]) - 3)
print(sigma_ost)
```

```python
f_ = sigma_fact / sigma_ost
f_
```

```python
f_table = 3.38
```

```python
if f_  > f_table:
    print('Различие между группами статистически значимое.' 
          'Следовательно есть различия среднего роста среди взрослых футболистов, хоккеистов и штангистов')
else:
    print('Нет статистически значимых различий.'
         'Следовательно нет различий среднего роста среди взрослых футболистов, хоккеистов и штангистов')
```
