### 1. Из колоды в 52 карты извлекаются случайным образом 4 карты. a) Найти вероятность того, что все карты – крести. б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

### Решение:
#### а)
#### $$\frac{52}{4} = 13\ -\ карт\ каждой\ масти$$
#### $$ P = \frac{13}{52} = \frac{1}{4} - вероятность\ достать\ из\ колоды\ 1\ крестовую\ карту$$
#### $$ P = \frac{12}{51} - вероятность\ достать\ из\ колоды\ 2\ крестовую\ карту$$
#### $$ P = \frac{11}{50} - вероятность\ достать\ из\ колоды\ 3\ крестовую\ карту$$
#### $$ P = \frac{10}{49} - вероятность\ достать\ из\ колоды\ 4\ крестовую\ карту$$
#### $$ P_{общ} = \frac{1}{4}\cdot\frac{12}{51}\cdot\frac{11}{50}\cdot\frac{10}{49} = \frac{66}{24990} - вероятность\ достать\ из\ колоды\ 4\ крестовых\ карты$$
#### б)
#### $$C_4^1 = \frac{4!}{1!\cdot3!} = 4\ -\ способа\ достать\ один\ туз$$
#### $$C_{32}^3 = \frac{32!}{3!\cdot29!} = 4960\ -\ способа\ достать\ 3\ произвольных\ карты\ из\ клолоды\ без\ тузов\ $$
#### $$ P_{общ} = C_4^1\cdot C_{32}^3 = 4*4960 =  19840- вероятность\ достать\ из\ колоды\ 4\ карты\ одна\ из\ которых\ туз$$

### 2. На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

### Решение:
#### $$C_10^{3} = \frac{10!}{3!(10!-3!)} = 120 -\ количество\ возможных\ комбинаций\ из\ 3\ кнопок $$
#### $$P = \frac{1}{120}\ -\ вероятность\ того,\ что\ человек,\ не\ знающий\ код,\ откроет\ дверь\ с\ первой\ попытки$$

### 3. В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

### Решение:
#### $$C_{15}^3 = \frac{15!}{3!(15!-3!)} = 455 -\ количество\ возможных\ комбинаций\ извлечь\ 3\ детали\ из\ ящика$$
#### $$C_9^3 = \frac{9!}{3!(9!-3!)} = 120 -\ количество\ возможных\ комбинаций\ извлечь\ 3\ окрашенных\ детали\ из\ 9\ окрашенных$$
#### $$P = \frac{C_9^3}{C_{15}^3} = \frac{120}{455} = \frac{24}{91}\ -\ вероятность\ того,\ что\ 3\ извлеченные\ детали\ окрашены$$


### 4. В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

### Решение:
#### $$C_2^2 = 1 -\ количество\ возможных\ комбинаций\ извлечь\ 2\ выиграшных\ билета\ из\ 2\ выигрышных$$
#### $$C_{100}^2 = 4950 -\ количество\ возможных\ комбинаций\ извлечь\ 2\ билета\ из\ 100$$
#### $$P = \frac{C_2^2}{C_{100}^2} = \frac{1}{4950}\ -\ вероятность\ того,\ что\ 2\ приобретенных\ билета\ окажутся\ выигрышными$$
